```python
import requests

url = 'https://search.rakuten.co.jp/search/mall/air+purifier/'

response = requests.get(url)
# print(response.text)
```

```python
# 웹페이지 파싱을 위한 패키지
from bs4 import BeautifulSoup

# response로 부터 html 가져오기
html = response.text

# BeautifulSoup 으로 html 파싱
soup = BeautifulSoup(html, 'html.parser')

# html에서 원하는 태그 찾기 - 상품 설명
get1 = soup.find_all('a', {"data-track-action":"title"})

# 텍스트만 추출
for info in get1:
    print(info.get_text(strip=True))   # 결과 출력

print('PRICE >>>')
# html에서 원하는 태그 찾기 - 가격
get2 = soup.find_all('div', {"class":"content description price"})

# 텍스트만 추출
for price in get2:
    print(price.get_text(strip=True))   # 결과 출력
```
